Q.1- What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?
- In hypothesis testing, Type I and Type II errors describe the two possible mistakes we can make when deciding based on sample data. They are crucial because they affect the reliability of statistical conclusions.

- Type I Error (False Positive)

Occurs when we reject a true null hypothesis (Hâ‚€).

In simple terms: you conclude something significant happened when it actually did not.

Symbol: Î± (alpha) â€” the significance level.

- Type II Error (False Negative)

Occurs when you fail to reject a false null hypothesis.

In simple terms: you miss a real effect.

Symbol: Î² (beta).

- Impact on Decision-Making

Choosing Î± (like 0.05 or 0.01) sets tolerance for Type I error.

Increasing sample size reduces both errors.

Decision-makers weigh the consequences:

Medicine â†’ avoid Type I error (donâ€™t approve ineffective drug).

Safety/diagnostics â†’ avoid Type II error (donâ€™t miss a real threat).

- The p-value is one of the most important concepts in hypothesis testing. It helps us decide whether to reject or not reject the null hypothesis (Hâ‚€).The p-value is the probability of obtaining a test statistic at least as extreme as the one observed, assuming the null hypothesis is true.

How to Interpret the P-value

1ï¸- If the p-value is small (â‰¤ significance level Î±)

2ï¸- If the p-value is large (> Î±)

The observed data is consistent with Hâ‚€.

We fail to reject the null hypothesis.

Conclusion: There is not enough evidence to support the alternative hypothesis.

Used when the population standard deviation (Ïƒ) is unknown.

Z-Test

Uses the t-distribution:

Looks similar to normal but has heavier tails.

- Use a Z-Test when:

Population standard deviation (Ïƒ) is known.

Sample size is large (n â‰¥ 30).

- Common examples:

Quality control

Large-scale surveys

- Use a T-Test when:

Sample size is small (n < 30).

- Common examples:

Comparing means from small samples

Testing effectiveness of a new drug on a small group

Q.5- Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?
- ANOVA (Analysis of Variance) is a statistical method used to compare three or more group means to determine whether at least one group differs significantly from the others.

Purpose of ANOVA

The main purpose of ANOVA is to test:

Are the means of three or more groups statistically different?

Instead of doing multiple t-tests (which increases Type I error), ANOVA allows a single overall test to check:
If ANOVA finds a significant difference, post-hoc tests (like Tukeyâ€™s test) identify exactly which groups differ.

- How ANOVA Extends Hypothesis Testing to More Than Two Groups

A t-test can only compare two means.

If you run multiple t-tests for multiple groups (A vs B, B vs C, C vs A), the chance of false positives increases.

ANOVA solves this by comparing all group means simultaneously using variance:

Between-group variance: how much the group means differ.

Within-group variance: how much individuals differ within each group.

ANOVA computes an F-statistic:

ð¹=Between-group variance/Within-group variance
	â€‹
A larger F means group means differ more relative to natural variability.



```python
#Q.6- Q.6- Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.
import numpy as np
from scipy.stats import norm

# -----------------------------
# Sample Data (You can replace)
# -----------------------------
data = np.array([12, 15, 14, 16, 13, 18, 17, 14, 15, 16])

population_mean = 14      # Hypothesized population mean (Î¼0)
population_std = 3        # Known population standard deviation (Ïƒ)
alpha = 0.05              # Significance level

# -----------------------------
# Z-test calculation
# -----------------------------
sample_mean = np.mean(data)
n = len(data)

# Z statistic
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# -----------------------------
# Visible output
# -----------------------------
print("Sample Mean:", sample_mean)
print("Population Mean (Hypothesized):", population_mean)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# -----------------------------
# Interpretation
# -----------------------------
if p_value < alpha:
    print("\nConclusion: Reject the null hypothesis (H0).")
    print("The sample mean is significantly different from the population mean.")
else:
    print("\nConclusion: Fail to reject the null hypothesis (H0).")
    print("There is no significant difference between the sample mean and the population mean.")
```

```python
#Q.8-Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python.
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Central Limit Theorem Simulation
# -----------------------------

np.random.seed(42)

# Step 1: Generate data from a non-normal distribution (Exponential)
population = np.random.exponential(scale=1.0, size=100000)

# Step 2: Draw many samples and compute sample means
sample_size = 50
num_samples = 1000

sample_means = []

for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# -----------------------------
# Plot 1: Histogram of Non-Normal Population
# -----------------------------
plt.figure(figsize=(6,4))
plt.hist(population, bins=50)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Non-Normal Distribution (Exponential)")
plt.show()

# -----------------------------
# Plot 2: Histogram of Sample Means (CLT)
# -----------------------------
plt.figure(figsize=(6,4))
plt.hist(sample_means, bins=30)
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.title("Distribution of Sample Means (CLT)")
plt.show()

sample_means[:5], population[:5]
```

```python
#Q.9 Q.9- Write a Python function to calculate and visualize the confidence interval for a sample mean.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def plot_mean_confidence_interval(data, confidence=0.95):
    """
    Calculate and visualize the confidence interval for the sample mean.
    """

    data = np.array(data)
    n = len(data)

    # Mean + Standard error
    mean = np.mean(data)
    std_err = stats.sem(data)

    # t-critical value
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n - 1)

    # CI bounds
    margin = t_crit * std_err
    lower = mean - margin
    upper = mean + margin

    print("Mean:", mean)
    print(f"{confidence*100:.1f}% Confidence Interval:", (lower, upper))

    # ---- Plot ----
    plt.figure(figsize=(6, 4))
    plt.errorbar(
        x=[0],
        y=[mean],
        yerr=[[mean - lower], [upper - mean]],
        fmt='o',
        capsize=10,
        linewidth=2
    )
    plt.xlim(-1, 1)
    plt.xticks([])
    plt.ylabel("Value")
    plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
    plt.grid(True, linestyle="--", alpha=0.5)

    plt.show()  # <-- ensures output displays

    return mean, lower, upper


# ---------------------------
# Example run
# ---------------------------
sample = [12, 15, 14, 16, 13, 18, 17, 14, 15, 16]
plot_mean_confidence_interval(sample)
```

q.10-  Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.


```python
from scipy.stats import chisquare

# Observed and expected frequencies
observed = [50, 30, 20]
expected = [40, 40, 20]

# Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square statistic:", chi_stat)
print("p-value:", p_value)
```

Meaning:

Chi-square statistic = 5.0
This measures how far the observed distribution deviates from expected.
A larger number means a larger mismatch.

p-value â‰ˆ 0.082
Compare this to Î± = 0.05 (a common significance level):

If p > 0.05 â†’ Fail to reject the null hypothesis
